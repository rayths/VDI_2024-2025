## 1. Tentukan data yang akan anda gunakan dalam visualisasi

```python
import pandas as pd

data_pengunjung  = pd.read_csv("/workspaces/VDI_2024-2025/Pertemuan 3/museum_visitors.csv")

data_pengunjung.head(10)
```

Data yang akan digunakan untuk visualisasi adalah data banyak pengunjung bulanan dari tahun 2014 hingga 2018 pada museum-museum yang terletak di Los Angeles, California yang terdiri dari Museum Avila Adobe, Firehouse Museum, Chinese American Museum, dan America Tropical Interpretive Center.

## 3. Buatlah visualisasi dengan mengunakan library python seperti matplotlib, seaborn atau tools lain

```python
# Import library
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates

# Mengubah tipe data 'Date' menjadi tipe data datetime
data_pengunjung['Date'] = pd.to_datetime(data_pengunjung['Date'])

# Membuat visualisasi
plt.figure(figsize=(12, 6))

# Membuat plot tren pengunjung setiap museum
plt.plot(data_pengunjung['Date'], data_pengunjung['Avila Adobe'], label='Avila Adobe', marker='o', color='#ff7f0e')
plt.plot(data_pengunjung['Date'], data_pengunjung['Firehouse Museum'], label='Firehouse Museum', marker='o', color='#d62728')
plt.plot(data_pengunjung['Date'], data_pengunjung['Chinese American Museum'], label='Chinese American Museum', marker='o', color='#1f77b4')
plt.plot(data_pengunjung['Date'], data_pengunjung['America Tropical Interpretive Center'], label='America Tropical Interpretive Center', marker='o', color='#2ca02c')

# Menampilkan format tanggal dua tingkat pada x-axis: Tahun dan Bulan
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_minor_locator(mdates.MonthLocator(bymonth=(3, 6, 9, 12)))

# Format untuk label tahun
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

# Format untuk label bulan
ax.xaxis.set_minor_formatter(mdates.DateFormatter('%b'))

# Membuat label tahun dan bulan dua tingkat
ax.tick_params(axis='x', which='major', pad=15, labelsize=12, rotation=0) 
ax.tick_params(axis='x', which='minor', pad=5, labelsize=10, rotation=0)  

# Judul dan label
plt.title('Tren Jumlah Pengunjung Bulanan Museum di Los Angeles (Januari 2014 - November 2018)', fontsize=16)
plt.xlabel('Tahun', fontsize=12)
plt.ylabel('Jumlah Pengunjung', fontsize=12)

# Menampilkan legenda
plt.legend(loc='upper right', fontsize=10)

# Menampilkan grid untuk membantu melihat tren tiap bulan dan tahun
plt.grid(True, linestyle='--', alpha=0.6)

# Menampilkan grafik dengan layout yang optimal
plt.tight_layout()
plt.show()
```

**Hasil:**\
Avila Adobe adalah museum yang paling populer, dengan jumlah pengunjung yang secara konsisten tinggi.
Firehouse Museum mengalami lonjakan pengunjung yang sangat signifikan pada September 2014, yang mungkin disebabkan oleh suatu acara khusus.
Museum-museum lainnya memiliki jumlah pengunjung yang lebih stabil dan tidak mengalami perubahan signifikan sepanjang tahun.
